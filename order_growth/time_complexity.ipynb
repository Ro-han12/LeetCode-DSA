```python
l=[1,2,3,4,5]
sum=0
for i in l :
    sum+=1
    print(f"The sum of 1 to {len(l)} is: {sum}")
    
product=1
for i in l:
    product*=product
    print(f"The product of {i} to {product} is: {product}")
#O(N) TIME COMPLEXITY
```

```python
def intostr(i):
    digits='0123456789'
    if i==0:
        return '0'
    res=""
    while i>0:
        res=digits[i%10]+res
        i=i//10 #USUALLY LOG
        return res 
intostr(123)
#LOG(N) TIME COMPLEXITY
```

```python
l=[1,23,45]
for i in range(0,len(l)//2):
    other=len(l)-i-1
    temp=l[i]
    l[i]=l[other]
    l[other]=temp 
print(l)
#O(N) TIME COMPLEXITY
```

```python
l=[98,99,100]
for i in range(0,len(l)//2):
    other=len(l)-i-1
    temp=l[i]
    l[i]=l[other]
    l[other]=temp
print(l)
#O(1) TIME COMPLEXITY
```

```python
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
fact(3)
#O(N) TIME COMPLEXITY
```

```python
def fib(n):
    if n==1 or n==0:
        return 1
    else:
        return fib(n-1)+fib(n-2)
fib(4)
#O(2^n) TIME COMPLEXITY
```

```python
def power(n):
    if n<1:
        return 0
    elif n==1:
        return 1
    else:
        prev=power(n//2)
        cur=prev*2
        print(cur)
        return cur
power(13)
#log(n) TIME COMPLEXITY
```

```python
def mod(a,b):
    if b<0:
        return -1 
    else:
        div=a//b 
        return a-div*2
mod(5,3)
#ONLY ARTHRMATIC OPERATIONS
#O(1) TIME COMPLEXITY
```

```python
def sum_digits(num):
    sum=0
    while(num>0):
        sum+=num%10
        num/=10
        return sum 
sum_digits(1234)
#log(n) TIME COMPLEXITY
```
